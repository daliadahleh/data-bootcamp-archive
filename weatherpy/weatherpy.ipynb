```python
# dependencies
from citipy import citipy
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from pprint import pprint
import json
import time
from api_keys import api_key
```

```python
# latitudes and longitudes ranges
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# initiate lists for lats/longs and cities
lats_longs = []
cities = []

# random combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
longs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lats_longs = zip(lats, longs)
```

```python
# find nearest cities
for lat_long in lats_longs:
    city = citipy.nearest_city(lat_long[0], lat_long[1]).city_name
    
    # only include city if we haven't seen it before
    if city not in cities:
        cities.append(city)
```

```python
len(cities)
```

### Perform API Calls


```python
weather={}

print('Beginning Data Retrieval\n-----------------------------')    

# start time
start = time.time()

for city in cities:
    print('Processing City # ' + str(cities.index(city)+1) + ' of ' + str(len(cities)-1) + ' | ' + city)
    
    response = requests.get('https://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=' + api_key + '&q=' + city)
    
    # only keep real responses
    if response.status_code == 200:
        weather[city] = response.json()
    else:
        print('City not found. Skipping..')
    
    # every 60 calls, pause until one minute has passed from the last start time (to avoid the 60 calls-per-minute limit)
    if (cities.index(city)+1) % 60 == 0:
        pause = 60 - (time.time()-start)
        print('Pausing for ' + str(round(pause)) + ' seconds to avoid reaching API limit..')
        time.sleep(pause)
        
        # reinitialize start time for next time we hit the max
        start = time.time()
    
print('-----------------------------\nData Retrieval Complete\n-----------------------------') 
```

### Convert Raw Data to DataFrame

```python
len(weather)
```

```python
test_item = list(weather.items())[0]
pprint(test_item)
```

```python
# initialize dataframe
weather_df = pd.DataFrame()
```

```python
# read into the df
for city in weather.keys():
   weather_df = weather_df.append(
       pd.DataFrame({
        'City': [city],
        'Cloudiness': [weather[city]['clouds']['all']],
        'Country': [weather[city]['sys']['country']],
        'Date': [weather[city]['dt']],
        'Humidity': [weather[city]['main']['humidity']],
        'Lat': [weather[city]['coord']['lat']],
        'Long': [weather[city]['coord']['lon']],
        'Max Temp': [weather[city]['main']['temp_max']],
        'Wind Speed': [weather[city]['wind']['speed']]
       })  
   )
```

```python
weather_df = weather_df.reset_index(drop=True)
weather_df.head()
```

```python
# save to csv
weather_df.to_csv('output/weather_csv.csv', index=False)
```
